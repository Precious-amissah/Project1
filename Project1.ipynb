```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('data.csv')
```

```python
data
```

```python
data.info()
```

```python
#finding the mean of the various attributes to replace the null values
mean_FA = data['fixed acidity'].mean()
mean_VA = data['volatile acidity'].mean()
mean_CA = data['citric acid'].mean()
mean_RS = data['residual sugar'].mean()
mean_C = data['chlorides'].mean()
mean_FSD = data['free sulfur dioxide'].mean()
mean_TSD = data['total sulfur dioxide'].mean()
mean_D = data['density'].mean()
mean_PH = data['pH'].mean()
mean_S = data['sulphates'].mean()
mean_A = data['alcohol'].mean()
```

```python
#Printing the mean values
print(f'mean_FA: {mean_FA}')
print(f'mean_VA: {mean_VA}')
print(f'mean_CA: {mean_CA}')
print(f'mean_RS: {mean_RS}')
print(f'mean_C: {mean_C}')
print(f'mean_FSD: {mean_FSD}')
print(f'mean_TSD: {mean_TSD}')
print(f'mean_D: {mean_D}')
print(f'mean_PH: {mean_PH}')
print(f'mean_S: {mean_S}')
print(f'mean_A: {mean_A}')
```

```python
#Filling the null values with the mean
data['fixed acidity']= data['fixed acidity'].fillna(mean_FA)
data['volatile acidity']= data['volatile acidity'].fillna(mean_FA)
data['citric acid']= data['citric acid'].fillna(mean_FA)
data['residual sugar']= data['residual sugar'].fillna(mean_FA)
data['chlorides']= data['chlorides'].fillna(mean_FA)
data['free sulfur dioxide']= data['free sulfur dioxide'].fillna(mean_FA)
data['total sulfur dioxide']= data['total sulfur dioxide'].fillna(mean_FA)
data['density']= data['density'].fillna(mean_FA)
data['pH']= data['pH'].fillna(mean_FA)
data['sulphates']= data['sulphates'].fillna(mean_FA)
data['alcohol']= data['alcohol'].fillna(mean_FA)
```
